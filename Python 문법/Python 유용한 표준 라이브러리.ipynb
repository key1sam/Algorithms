### 내장함수
 - 기본 입출력 함수부터 정렬 함수까지 기본적인 함수들을 제공합니다.
 - 파이썬 프로그램을 작성할 때 없어서는 안 되는 필수적인 기능을 포함하고 있습니다.

```python
# sum ()
result = sum([1, 2, 3, 4, 5]) # 리스트, 튜플 등의 합을 반환
print(result)

# min(), max() 
min_result = min(9,8,7,6) # 리스트에서 가장 작은 값 반환
max_result = max(9,8,7,6) # 리스트에서 가장 큰 값 반환
print(min_result, max_result)

# eval() # 문자열 수식으로 된 값을 계산한 후 숫자로 바꾸어 출력
result = eval("(3+5)*7")
print(result)

# sorted()
result = sorted([9, 1, 8, 5, 4])
reverse_result = sorted([9, 1, 8, 5, 4], reverse = True)
print(result)
print(reverse_result)

# sorted() with key
array = [('홍길동', 35), ('이순신', 75), ('아무개', 50)]
result = sorted(array, key=lambda x:x[1], reverse = True)
print(result)
# sorted에서 key를 정해줄 수 있다. lambda로 지정을 많이해주며 
#x[1]인 35, 75, 50 등을 기준으로 정렬. reverse = True를 통해 x[1]값이 큰 값을 기준으로 정렬을 수행한다.
                                      
```

### itertools
 - 파이썬에서 반복되는 형태의 데이터를 처리하기 위한 유용한 기능들을 제공
 - 특히 순열과 조합 라이브러리는 코딩테스트에서 자주 사용됩니다.

> #### 순열과 조합
> - 모든 경우의 수를 고려해야 할 때 어떤 라이브러리를 효과적으로 사용할 수 있을 >것>인가?
> - 순열 : 서로 다른 n개에서 서로 다른 r개를 선택하여 일렬로 나열하는 것
>   - {'A', 'B', 'C'}에서 세 개를 선택하여 나열 : 'ABC', 'ACB', 'BAC', 'BCA', 'CAB', 'CBA'
> - 조합 : 서로 다른 n개에서 순서에 상관 없이 서로 다른 r개를 선택하는 것
>   - {'A', 'B', 'C'}에서 순서를 고려하지 않고 두 개를 뽑는 경우 : 'AB', 'AC', 'BC'

>

```python
# 순열
from itertools import permutations #itertools에서 순열과 관련된 함수 가져오기
data = ['A', 'B', 'C']
result = list(permutations(data, 3))# data에서 3개를 골라서 조합을 나열한다.
print(result)
```

```python
# 조합
from itertools import combinations #itertools에서 조합과 관련된 함수 가져오기
data = ['A', 'B', 'C']
result = list(combinations(data, 2))# data에서 2개를 골라서 조합을 나열한다.
print(result)
```

```python
# 중복순열
from itertools import product
data = ['A', 'B' ,'C']
result = list(product(data, repeat = 2)) #2개를 뽑는 모든 순열 구하기 (중복 허용)
print(result)
```

```python
# 중복 조합
from itertools import combinations_with_replacement
data = ['A', 'B', 'C']
result = list(combinations_with_replacement(data, 2)) #2개를 뽑는 모든 조합 구하기 (중복 허용)
print(result)
```

### collections
 - 덱(deque), 카운터(Counter) 등의 유용한 자료구조를 제공합니다.

>#### Counter
>- 파이썬 collections 라이브러리의 Counter는 등장 횟수를 세는 기능을 제공합니다.
>- 리스트와 같은 반복 가능한(itertable) 객체가 주어졌을 때 내부의 원소가 몇 번씩 등장했는지 알려줍니다.

```python
from collections import Counter

counter = Counter(['Red', 'blue', 'red', 'green', 'blue', 'blue'])
print(counter['blue'])
print(counter['green'])
print(dict(counter)) # 원소의 개수를 사전 자료형으로 반환
```

### math
 - 필수적인 수학적 기능을 제공합니다.
 - 팩토리얼, 제곱근, 최대공약수(GCD), 삼국함수 관련 함수부터 파이(Pi)와 같은 상수 포함

> #### 최대 공약수와 최소 공배수
> 최대 공약수는 math 라이브러리의 gcd() 함수를 이용할 수 있습니다.

```python
import math

# 최소 공배수 (LCM)를 구하는 함수
def lcm(a, b) :
    return a*b // math.gcd(a,b) # // 는 몫을 출력해주는 연산자. python에서 /는 float으로 출력
a = 21
b= 14
print(math.gcd(21, 14)) # 최대 공약수(GCD) 계산
print(lcm(21, 14)) # 최소 공배수(LCM) 계산
```
